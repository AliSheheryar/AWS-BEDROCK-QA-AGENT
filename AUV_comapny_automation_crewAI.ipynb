## Kicking off the Crew

```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool, ScrapeWebsiteTool, WebsiteSearchTool
import openai
import os

# Set OpenAI and other necessary API keys
openai.api_key = "your-openai-api-key"
os.environ['OPENAI_MODEL_NAME'] = 'gpt-4o-mini'

# **Agent 1: Perplexity-based AUV Manufacturer Search**
perplexity_search_agent = Agent(
    role="PerplexitySearchAgent",
    goal="Find a list of companies manufacturing Autonomous Underwater Vehicles (AUVs) using Perplexity.",
    backstory="This agent uses Perplexity to search for and list companies that manufacture Autonomous Underwater Vehicles (AUVs). It gathers relevant information on companies from the Perplexity database.",
    tools=[SerperDevTool()],
    verbose=True
)

# **Agent 2: ChatGPT-based AUV Manufacturer Search**
chatgpt_search_agent = Agent(
    role="ChatGPTSearchAgent",
    goal="Find a list of companies manufacturing Autonomous Underwater Vehicles (AUVs) using ChatGPT.",
    backstory="This agent uses ChatGPT to search for and list companies that manufacture Autonomous Underwater Vehicles (AUVs). It queries the model to generate relevant company data.",
    tools=[WebsiteSearchTool()],
    verbose=True
)

# **Agent 3: Clean and Combine Company List**
clean_list_agent = Agent(
    role="ListCleaner",
    goal="Combine the results from Perplexity and ChatGPT agents and remove duplicates.",
    backstory="This agent takes the lists from both the Perplexity and ChatGPT search agents, combines them, and removes any duplicate entries to create a unified list of companies.",
    verbose=True
)

# **Agent 4: Research Topics for Each Company**
research_agent = Agent(
    role="ResearchAgent",
    goal="Run detailed research on each company for technical specifications, sensor suite, and other features.",
    backstory="This agent conducts detailed research on each company by querying both Perplexity and ChatGPT to gather in-depth information about technical specifications, sensor suites, applications, and other relevant data.",
    tools=[SerperDevTool(), WebsiteSearchTool()],
    verbose=True
)

# **Agent 5: Google Docs Document Creation**
doc_creation_agent = Agent(
    role="DocCreationAgent",
    goal="Generate Google Docs for each company with the required format.",
    backstory="This agent creates Google Docs for each company by formatting and populating them with detailed information gathered during the research phase. It also includes images fetched from web scraping.",
    tools=[ScrapeWebsiteTool()],
    verbose=True
)

# **Agent 6: Google Spreadsheet Creation**
spreadsheet_creation_agent = Agent(
    role="SpreadsheetAgent",
    goal="Create a Google Spreadsheet with company data.",
    backstory="This agent generates a Google Spreadsheet where each row represents a company and the columns contain the research data such as technical specifications, sensor suite, and operational features.",
    tools=[SerperDevTool()],
    verbose=True
)

# **Task 1: Fetch Companies**
fetch_companies_task = Task(
    description="Run the same prompt in different AI tools (Perplexity and ChatGPT) to find companies manufacturing AUVs.",
    agent=perplexity_search_agent,
    expected_output="List of companies manufacturing AUVs."
)

# **Task 2: Clean and Combine List**
clean_list_task = Task(
    description="Clean and combine the lists from both agents, remove duplicates, and format them.",
    agent=clean_list_agent,
    expected_output="Unified list of companies."
)

# **Task 3: Research Topics**
research_topics_task = Task(
    description="Research detailed topics for each company (technical specs, sensor suite, etc.) using ChatGPT and Perplexity.",
    agent=research_agent,
    expected_output="Detailed research for each company."
)

# **Task 4: Document Creation**
doc_creation_task = Task(
    description="Generate Google Docs with detailed information for each company, including images.",
    agent=doc_creation_agent,
    expected_output="Google Docs for each company."
)

# **Task 5: Create Spreadsheet**
spreadsheet_creation_task = Task(
    description="Create a Google Spreadsheet with company data in the specified format.",
    agent=spreadsheet_creation_agent,
    expected_output="Google Spreadsheet with company data."
)

# **Task 6: Re-research for Missing Information**
re_search_task = Task(
    description="Re-research for missing information in the Google Docs or Spreadsheet for each company.",
    agent=research_agent,
    expected_output="Completed and filled-in data."
)

# **Create Crew**
crew = Crew(
    agents=[perplexity_search_agent, chatgpt_search_agent, clean_list_agent, research_agent, doc_creation_agent, spreadsheet_creation_agent],
    tasks=[fetch_companies_task, clean_list_task, research_topics_task, doc_creation_task, spreadsheet_creation_task, re_search_task],
    verbose=True
)

# **Kick off the Crew to start processing**
result = crew.kickoff(inputs={"prompt": "List companies manufacturing Autonomous Underwater Vehicles"})

# **Outputs**: The result will include the Google Docs and Spreadsheet, ready for further action.
print(result)
```
