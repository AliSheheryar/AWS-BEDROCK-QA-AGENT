## Creating Structured Output

```python
from pydantic import BaseModel, Field
from typing import List

class SocialMediaPost(BaseModel):
    platform: str = Field(..., description="The social media platform where the post will be published (e.g., Twitter, LinkedIn).")
    content: str = Field(..., description="The content of the social media post, including any hashtags or mentions.")

class ContentOutput(BaseModel):
    article: str = Field(..., description="The article, formatted in markdown.")
    social_media_posts: List[SocialMediaPost] = Field(..., description="A list of social media posts related to the article.")
```

## Loading Tasks and Agents YAML files

```python
# Define file paths for YAML configurations
files = {
    'agents': 'config/agents.yaml',
    'tasks': 'config/tasks.yaml'
}

# Load configurations from YAML files
configs = {}
for config_type, file_path in files.items():
    with open(file_path, 'r') as file:
        configs[config_type] = yaml.safe_load(file)

# Assign loaded configurations to specific variables
agents_config = configs['agents']
tasks_config = configs['tasks']
```

## Creating Crew, Agents, and Tasks

```python
# Creating Agents
market_news_monitor_agent = Agent(
    config=agents_config['market_news_monitor_agent'],
    tools=[SerperDevTool(), ScrapeWebsiteTool()],
    llm=groq_llm,
)

data_analyst_agent = Agent(
    config=agents_config['data_analyst_agent'],
    tools=[SerperDevTool(), WebsiteSearchTool()],
    llm=groq_llm,
)

content_creator_agent = Agent(
    config=agents_config['content_creator_agent'],
    tools=[SerperDevTool(), WebsiteSearchTool()],
)

quality_assurance_agent = Agent(
    config=agents_config['quality_assurance_agent'],
)

# Creating Tasks
monitor_financial_news_task = Task(
    config=tasks_config['monitor_financial_news'],
    agent=market_news_monitor_agent
)

analyze_market_data_task = Task(
    config=tasks_config['analyze_market_data'],
    agent=data_analyst_agent
)

create_content_task = Task(
    config=tasks_config['create_content'],
    agent=content_creator_agent,
    context=[monitor_financial_news_task, analyze_market_data_task]
)

quality_assurance_task = Task(
    config=tasks_config['quality_assurance'],
    agent=quality_assurance_agent,
    output_pydantic=ContentOutput
)

# Creating Crew
content_creation_crew = Crew(
    agents=[
        market_news_monitor_agent,
        data_analyst_agent,
        content_creator_agent,
        quality_assurance_agent
    ],
    tasks=[
        monitor_financial_news_task,
        analyze_market_data_task,
        create_content_task,
        quality_assurance_task
    ],
    verbose=True
)
```

## Kicking off the Crew

```python
from pydantic import BaseModel
from typing import List, Dict
from IPython.display import display, Markdown
from crewai import Agent, Task, Crew
from crewai_tools import ScrapeWebsiteTool, WebsiteSearchTool
import openai
import os

# Set OpenAI API key and environment
openai.api_key = "your-openai-api-key"
os.environ['OPENAI_MODEL_NAME'] = 'gpt-4o-mini'
groq_llm = "groq/llama-3.3-70b-versatile"

# **Pydantic Models to Structure the Output**

class CompanyData(BaseModel):
    company_name: str
    product_name: str
    description: str
    features: str
    recent_news: str
    address: str
    website: str

class FinalResult(BaseModel):
    article: List[CompanyData]

# **Define Agents**

# Agent 1: groq/llama-3.3-70b AUV Manufacturer Search
groq_search_agent = Agent(
    role="GroqSearchAgent",
    goal="Find a list of companies manufacturing Autonomous Underwater Vehicles (AUVs) using Groq/llama-3.3-70b.",
    backstory="This agent uses the Groq/llama-3.3-70b model to search for and list companies that manufacture Autonomous Underwater Vehicles (AUVs).",
    tools=[WebsiteSearchTool()],
    llm=groq_llm,
    verbose=True
)

# **Agent 2: ChatGPT-based AUV Manufacturer Search**
chatgpt_search_agent = Agent(
    role="ChatGPTSearchAgent",
    goal="Find a list of companies manufacturing Autonomous Underwater Vehicles (AUVs) using ChatGPT.",
    backstory="This agent uses ChatGPT to search for and list companies that manufacture Autonomous Underwater Vehicles (AUVs).",
    tools=[WebsiteSearchTool()],
    llm=groq_llm,
    verbose=True
)

# **Agent 3: Clean and Combine Company List**
clean_list_agent = Agent(
    role="ListCleaner",
    goal="Combine the results from Groq/llama-3.3-70b and ChatGPT agents and remove duplicates.",
    backstory="This agent takes the lists from both the Groq/llama-3.3-70b and ChatGPT search agents, combines them, and removes any duplicate entries to create a unified list of companies.",
    verbose=True
)

# **Agent 4: Research Topics for Each Company**
research_agent = Agent(
    role="ResearchAgent",
    goal="Run detailed research on each company for technical specifications, sensor suite, and other features.",
    backstory="This agent conducts detailed research on each company by querying both Groq/llama-3.3-70b and ChatGPT to gather in-depth information about technical specifications, sensor suites, applications, and other relevant data.",
    tools=[ScrapeWebsiteTool(), WebsiteSearchTool()],
    llm=groq_llm,
    verbose=True
)

# **Agent 5: Google Docs Document Creation**
doc_creation_agent = Agent(
    role="DocCreationAgent",
    goal="Generate Google Docs for each company with the required format.",
    backstory="This agent creates Google Docs for each company by formatting and populating them with detailed information gathered during the research phase. It also includes images fetched from web scraping.",
    tools=[ScrapeWebsiteTool()],
    verbose=True
)

# **Agent 6: Google Spreadsheet Creation**
spreadsheet_creation_agent = Agent(
    role="SpreadsheetAgent",
    goal="Create a Google Spreadsheet with company data.",
    backstory="This agent generates a Google Spreadsheet where each row represents a company and the columns contain the research data such as technical specifications, sensor suite, and operational features.",
    tools=[SerperDevTool()],
    verbose=True
)

# **Task 1: Fetch Companies**
fetch_companies_task = Task(
    description="Run the same prompt in different AI tools (Groq/llama-3.3-70b and ChatGPT) to find companies manufacturing AUVs.",
    agent=groq_search_agent,
    expected_output="List of companies manufacturing AUVs."
)

# **Task 2: Clean and Combine List**
clean_list_task = Task(
    description="Clean and combine the lists from both agents, remove duplicates, and format them.",
    agent=clean_list_agent,
    expected_output="Unified list of companies."
)

# **Task 3: Research Topics**
research_topics_task = Task(
    description="Research detailed topics for each company (technical specs, sensor suite, etc.) using Groq/llama-3.3-70b and ChatGPT.",
    agent=research_agent,
    expected_output="Detailed research for each company."
)

# **Create Crew**
crew = Crew(
    agents=[groq_search_agent, chatgpt_search_agent, clean_list_agent, research_agent],
    tasks=[fetch_companies_task, clean_list_task, research_topics_task],
    verbose=True
)

# **Function to Continuously Fetch Data and Format as Pydantic Model**
def fetch_and_store_data():
    companies_data = []

    # Loop to continually fetch data (adjust the range as needed)
    for _ in range(5):  # Example: Run for 5 iterations, adjust as needed
        try:
            # Kick off the crew to gather companies
            result = crew.kickoff(inputs={"prompt": "List companies manufacturing Autonomous Underwater Vehicles"})

            # Check if result is valid and contains the expected data
            if result is not None and hasattr(result, 'pydantic'):
                fetched_companies = result.pydantic.dict().get('companies', [])
                if fetched_companies:
                    for company in fetched_companies:
                        # Create Pydantic objects for each company
                        companies_data.append(CompanyData(
                            company_name=company.get("company_name", ""),
                            product_name=company.get("product_name", ""),
                            description=company.get("description", ""),
                            features=company.get("features", ""),
                            recent_news=company.get("recent_news", ""),
                            address=company.get("address", ""),
                            website=company.get("website", "")
                        ))
                else:
                    print("No companies found in this iteration.")
            else:
                print("Error: No result returned or result format is incorrect.")

        except Exception as e:
            print(f"Error during data fetching: {e}")

    # Wrap the company data into the FinalResult Pydantic model
    final_result = FinalResult(article=companies_data)

    # Output the final result using Pydantic's dict() method
    display(Markdown(format_companies_to_markdown(final_result.dict()['article'])))

# **Function to Convert Companies Data to Markdown Format**
def format_companies_to_markdown(companies: List[CompanyData]) -> str:
    markdown_str = "| Company Name       | Product Name      | Description                                                             | Features                                  | Recent News Summary                                                                       | Address                                 | Website                        |\n"
    markdown_str += "|--------------------|-------------------|-------------------------------------------------------------------------|-------------------------------------------|-----------------------------------------------------------------------------------------|-----------------------------------------|-------------------------------|\n"

    for company in companies:
        markdown_str += f"| {company.company_name} | {company.product_name} | {company.description} | {company.features} | {company.recent_news} | {company.address} | {company.website} |\n"

    return markdown_str

# **Run the Data Fetch and Output Results**
fetch_and_store_data()
```
