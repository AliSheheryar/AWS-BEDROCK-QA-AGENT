## Kicking off the Crew

```python
# Install only once
# !pip install crewai crewai-tools openai python-dotenv --quiet

import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool

# === Tool: Real-time Google Search ===
search_tool = SerperDevTool()

# === Agent 1: Discovery Agent ===
rov_discovery_agent = Agent(
    role="ROV Discovery Expert",
    goal="Find all companies manufacturing ROVs and list their ROV products",
    backstory=(
        "You're an expert in online research. Your job is to find every manufacturer of Remotely Operated "
        "Vehicles (ROVs) in the world by querying search engines and analyzing sources. You iterate on prompts, "
        "try different keywords and combinations until the list is exhaustive."
    ),
    verbose=True,
    memory=True,
    tools=[search_tool]
)

# === Agent 2: Research Agent ===
rov_research_agent = Agent(
    role="ROV Researcher",
    goal="Deeply research every discovered ROV company and their individual ROV products",
    backstory=(
        "You're a thorough researcher trained to extract detailed, structured, and factual data "
        "about ROV companies and each of their ROV products including sensors, capabilities, specs, and control systems."
    ),
    verbose=True,
    memory=True
)

# === Agent 3: Report Writer ===
rov_report_writer = Agent(
    role="ROV Markdown Report Generator",
    goal="Format all ROV company and product data into clean, professional markdown files",
    backstory="You specialize in writing structured reports with headers, subheaders and rich product breakdowns in markdown format.",
    verbose=True,
    memory=False
)

# === Agent 4: Auditor ===
rov_auditor_agent = Agent(
    role="Report Completeness Checker",
    goal="Ensure all markdown reports contain the required sections and no topic is skipped",
    backstory="You're the last step of QA before delivery. You check reports for completeness, missing products, or weak sections.",
    verbose=True,
    memory=False
)

# === Task 1: Discover ROV Companies ===
task_discover_rov_companies = Task(
    description=(
        "Use the internet (Google, directories, Wikipedia, forums, vendor listings) to create a comprehensive list "
        "of all companies manufacturing Remotely Operated Vehicles (ROVs). "
        "Include military, scientific, offshore, and commercial vendors. "
        "Try different search terms and improve prompts. Return a markdown bullet list with company names."
    ),
    expected_output="A markdown list of unique ROV company names (bullet points).",
    tools=[search_tool],
    agent=rov_discovery_agent
)

# === Task 2: Research each company and its ROV products ===
task_research_rov_products = Task(
    description=(
        "For each ROV company from the list, find detailed information about:\n"
        "1. Company overview and location\n"
        "2. List of ROV products they make\n"
        "3. For each ROV product:\n"
        "   - Name and purpose\n"
        "   - Depth rating, size, weight\n"
        "   - Sensors included\n"
        "   - Control system and data links\n"
        "   - Power system\n"
        "   - Cameras or imaging tools\n"
        "   - Notable deployments or customers\n"
        "   - Pictures or public references\n"
    ),
    expected_output="A structured dictionary of detailed company + ROV product data.",
    agent=rov_research_agent
)

# === Task 3: Markdown Generation ===
task_generate_markdown = Task(
    description=(
        "Turn the research into a markdown file per company. Each file should contain:\n"
        "- Company name\n"
        "- Company overview\n"
        "- A section for each ROV product with full technical details\n"
        "Use clear headers (##, ###) and bullet points."
    ),
    expected_output="Markdown files saved per company with full ROV product detail.",
    agent=rov_report_writer
)

# === Task 4: Audit (optional) ===
task_audit_reports = Task(
    description=(
        "Review each markdown report and verify that all required sections are included.\n"
        "Ensure no ROV products are skipped or missing key specs."
    ),
    expected_output="Summary of completeness issues, if any.",
    agent=rov_auditor_agent
)

# === Assemble the Crew ===
crew = Crew(
    agents=[
        rov_discovery_agent,
        rov_research_agent,
        rov_report_writer,
        rov_auditor_agent
    ],
    tasks=[
        task_discover_rov_companies,
        task_research_rov_products,
        task_generate_markdown,
        task_audit_reports
    ],
    process=Process.sequential
)

# === Kick Off the Workflow ===
result = crew.kickoff()
print("\n✅ FINAL ROV RESEARCH OUTPUT ✅\n")
print(result)
```
